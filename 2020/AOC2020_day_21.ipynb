```
test_input = """mxmxvkd kfcds sqjhc nhms (contains dairy, fish)
trh fvjkl sbzzf mxmxvkd (contains dairy)
sqjhc fvjkl (contains soy)
sqjhc mxmxvkd sbzzf (contains fish)""".split('\n')

import copy
menu = []
ingredients = dict()
ing_num = dict()
for i in test_input:
  food, aler = i.split(' (contains ')
  #print (food,aler[0:len(aler-1)])
  menu.append (((food.split (' ')),(aler[0:len(aler)-1].split(', '))))
print ('menu:',*menu,sep=('\n'))
ori_menu = copy.deepcopy(menu)
result =dict()
count =0
while True:
  for i in menu:
    for k in i[0]:
      if k not in ingredients.keys():
        ingredients [k] = i[1]
      else:
        vtest=list(ingredients[k])
        for j in i[1]:
          vtest.append(j)
        ingredients [k] =vtest

  for i in ingredients.keys():
    vc= 0
    for j in menu:
      if i in j[0]:
        vc+=1
    ing_num[i]=vc
  #print('ingredients:')
  #for key, value in ingredients.items():
      #print(key, ':', value,ing_num[key])
  #print ('\n')
  for i in ingredients:
    freq={}
    if ing_num [i]> 1:
      for x in ingredients[i]: 
        if (x in freq): 
            freq[x] += 1
        else: 
            freq[x] = 1
      max=0
      for j in freq:
        if freq[j] > max:
          max = freq[j]
      cmax = 0
      for j in freq:
        #print (j,freq[j])
        if freq[j] == max:
          cmax +=1
      if cmax == 1:
        result [i] = [x for x in freq if freq[x] == max][0]
  #print (result)
  for i in result:
    for j,k in enumerate(menu):
      if i in k[0]:
        menu[j][0].remove(i)
      if result[i] in k[1]:
        menu[j][1].remove(result[i])
  #clean menu
  for i,j in enumerate(menu):
    if len(j[1]) == 0:
      menu.pop(i)
  #print (menu)
  if len(menu) == 0:
    break
  count +=1

print ('\n',result)

sol1=0
for i in ori_menu:
  for j in i[0]:
    if j not in result:
      sol1+=1
print (sol1)
```
