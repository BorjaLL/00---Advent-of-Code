The safe has a dial with only an arrow on it; around the dial are the numbers 0 through 99 in order. As you turn the dial, it makes a small click noise as it reaches each number.

The attached document (your puzzle input) contains a sequence of rotations, one per line, which tell you how to open the safe. A rotation starts with an L or R which indicates whether the rotation should be to the left (toward lower numbers) or to the right (toward higher numbers). Then, the rotation has a distance value which indicates how many clicks the dial should be rotated in that direction.

So, if the dial were pointing at 11, a rotation of R8 would cause the dial to point at 19. After that, a rotation of L19 would cause it to point at 0.

Because the dial is a circle, turning the dial left from 0 one click makes it point at 99. Similarly, turning the dial right from 99 one click makes it point at 0.

So, if the dial were pointing at 5, a rotation of L10 would cause it to point at 95. After that, a rotation of R5 could cause it to point at 0.

The dial starts by pointing at 50.

You could follow the instructions, but your recent required official North Pole secret entrance security training seminar taught you that the safe is actually a decoy. The actual password is the number of times the dial is left pointing at 0 after any rotation in the sequence.

```python
def fInput(vtest):
    sample = '''L68
L30
R48
L5
R60
L55
L1
L99
R14
L82'''
    if vtest == 1:
        # Make sure this path is correct for your system
        with open('/Users/bllanderas/Library/CloudStorage/OneDrive-Deloitte(O365D)/Documents/VSCODE/00---Advent-of-Code/2025/data/day1.txt', 'r') as f:
            data = f.read()
    else:
        data = sample
    lines = [line.strip() for line in data.splitlines() if line.strip()]
    return [(line[0], int(line[1:])) for line in lines]

def zero_passes(start_pos, dir_char, steps):
    """
    Calculates the number of times the dial points at 0 during a rotation.
    
    Args:
        start_pos (int): The current position of the dial (0-99).
        dir_char (str): The direction of rotation ('L' for left, 'R' for right).
        steps (int): The number of steps to rotate.
        
    Returns:
        int: The number of times the dial points at 0 during this rotation.
    """
    
    # Normalize start_pos to be within 0-99
    s = start_pos % 100
    
    # If starting at zero, the next distinct zero hit is 100 steps away.
    # We don't count the starting position itself as a "hit during rotation".
    if s == 0:
        first_hit_distance = 100
    else:
        if dir_char == 'R':
            # Distance to reach 0 when rotating right: (100 - current_pos) % 100
            # If current_pos is 50, distance to 0 is 50.
            first_hit_distance = (100 - s) % 100
        else:  # 'L'
            # Distance to reach 0 when rotating left: current_pos
            # If current_pos is 50, distance to 0 is 50.
            first_hit_distance = s
            
        # If first_hit_distance is 0, it means s was 0, but we handled that above.
        # This ensures that if s was 0, first_hit_distance is 100.
        if first_hit_distance == 0:
            first_hit_distance = 100

    # If the total steps are less than the distance to the first 0, no hits occur.
    if steps < first_hit_distance:
        return 0
    
    # Otherwise, at least one hit occurs (at first_hit_distance),
    # plus one hit for every subsequent full 100-step cycle.
    return 1 + (steps - first_hit_distance) // 100

def apply_moves(moves, start=50):
    """
    Applies a series of rotations to the dial and counts total zero hits.
    
    Args:
        moves (list): A list of tuples, where each tuple is (direction_char, steps).
        start (int): The initial position of the dial (default is 50).
        
    Returns:
        int: The total number of times the dial pointed at 0.
    """
    pos = start
    total_zero_hits = 0
    
    for dir_char, steps in moves:
        # Calculate zero passes for the current move
        total_zero_hits += zero_passes(pos, dir_char, steps)
        
        # Update the dial's final position after the move
        if dir_char == 'R':
            pos = (pos + steps) % 100
        else: # 'L'
            pos = (pos - steps) % 100
            # Handle negative results from modulo for Python (e.g., -5 % 100 is -5, not 95)
            if pos < 0:
                pos += 100
                
    return total_zero_hits

# --- Main execution ---

# Example usage with the sample input (vtest=0)
print("Running with sample input:")
sample_moves = fInput(1)
result_sample = apply_moves(sample_moves, start=50)
print(f"Total zero hits for sample: {result_sample}") # Expected: 6

# To run with your file input (vtest=1)
# Uncomment the following lines when you want to use your actual data file
# print("Running with file input:")
# file_moves = fInput(1)
# result_file = apply_moves(file_moves, start=50)
# print(f"Total zero hits for file: {result_file}")
```
