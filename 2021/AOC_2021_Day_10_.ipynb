```python
test_input = """[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]"""
chunks = [i for i in test_input.split('\n')]
```

```python
def checkLine (line):
  openers=[]
  for i in line:#range (0, len(line)):
    if i in ['(','[','{','<']:
      openers.append(i)
      #print (openers)
    else:
      if i == ')' and openers[-1] == '(':
        openers.pop(-1)
      elif i == ']' and openers[-1] == '[':
        openers.pop(-1)
      elif i == '}' and openers[-1] == '{':
        #print (i, openers[-1])
        openers.pop(-1)
      elif i == '>' and openers[-1] == '<':
        openers.pop(-1)
      else: 
        #print ("found ", i, "expected matching ", openers[-1])
        return i
    #print (openers)

def part1 (chunks):
  illegal = 0
  for i in chunks:
    char=checkLine(i)
    if char == ')':
      illegal += 3
    elif char == ']':
      illegal += 57
    elif char == '}':
      illegal += 1197
    elif char == '>':
      illegal += 25137
  print (illegal)
#checkLine('{([(<{}[<>[]}>{[]{[(<()>')
part1(chunks)
```
