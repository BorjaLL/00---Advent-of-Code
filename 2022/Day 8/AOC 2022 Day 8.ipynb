Transform input into map

```python
forest = []
for i, j in enumerate(input):
    size = len(j)
    forest.append([])
    for k in range(size):
        forest[i].append(int(j[k]))

# forest
# [[3, 0, 3, 7, 3],
#  [2, 5, 5, 1, 2],
#  [6, 5, 3, 3, 2],
#  [3, 3, 5, 4, 9],
#  [3, 5, 3, 9, 0]]
```

Part 1

```python
sol1 = 0
tall = len(forest) - 1
wide = len(forest[0]) - 1


def visCheck(y, x):
    global forest
    tree = forest[y][x]
    if y == 0 or x == 0 or y == tall or x == wide:
        return 1
    else:
        # UP
        checkx = x
        checky = y-1
        while checky >= 0:
            if checky == 0 and tree > forest[checky][checkx]:
                return 1
            elif tree <= forest[checky][checkx]:
                break
            else:
                checky -= 1
        # DOWN
        checkx = x
        checky = y+1
        while checky <= tall:
            if checky == tall and tree > forest[checky][checkx]:
                return 1
            elif tree <= forest[checky][checkx]:
                break
            else:
                checky += 1
        # LEFT
        checkx = x-1
        checky = y
        while checkx >= 0:
            if checkx == 0 and tree > forest[checky][checkx]:
                return 1
            elif tree <= forest[checky][checkx]:
                break
            else:
                checkx -= 1
        # RIGHT
        checkx = x+1
        checky = y
        while checkx <= wide:
            if checkx == wide and tree > forest[checky][checkx]:
                return 1
            elif tree <= forest[checky][checkx]:
                break
            else:
                checkx += 1
    return 0


for y in range(tall+1):
  for x in range(wide+1):
    var = visCheck (y,x)
    #print (y,x,var)
    sol1 += var

# print(visCheck(2, 1))

sol1
```

```python
sol2 = 0
tall = len(forest) - 1
wide = len(forest[0]) - 1


def scenCheck(y, x):
    global forest
    tree = forest[y][x]
    score = 1
    # UP
    scoreup = 0
    checkx = x
    checky = y - 1
    while checky >= 0:
        if tree > forest[checky][checkx]:
            scoreup += 1
            checky -= 1
        elif tree <= forest[checky][checkx]:
            scoreup += 1
            break
    # DOWN
    scoredown = 0
    checkx = x
    checky = y + 1
    while checky <= tall:
        if tree > forest[checky][checkx]:
            scoredown += 1
            checky += 1
        elif tree <= forest[checky][checkx]:
            scoredown += 1
            break
    # LEFT
    scoreleft = 0
    checkx = x - 1
    checky = y
    while checkx >= 0:
        if tree > forest[checky][checkx]:
            scoreleft += 1
            checkx -= 1
        elif tree <= forest[checky][checkx]:
            scoreleft += 1
            break
    # RIGHT
    scoreright = 0
    checkx = x + 1
    checky = y
    while checkx <= wide:
        if tree > forest[checky][checkx]:
            scoreright += 1
            checkx += 1
        elif tree <= forest[checky][checkx]:
            scoreright += 1
            break
    # print(x, y, scoreup, scoreleft, scoreright, scoredown)
    score = scoreup * scoredown * scoreleft * scoreright
    return score


for y in range(tall + 1):
    for x in range(wide + 1):
        var = scenCheck(y, x)
        if var > sol2:
            sol2 = var
# y = 1
# x = 2
# if scenCheck(y, x) != 4:
#     print("test", y, ",", x, "KO")
# y = 3
# x = 2
# if scenCheck(y, x) != 8:
#     print("test", y, ",", x, "KO")



sol2
```
